# Lable Encoding

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
print(df.month.value_counts())
```

```python
df.month=le.fit_transform(df.month)
print(df.month.value_counts())
```

```python
df.head()
```

# Preprocessing the Data

```python
#df.Address=le.fit_transform(df.Address)
df.X=le.fit_transform(df.X)
df.Y=le.fit_transform(df.Y)
df.month=le.fit_transform(df.month)
df.temp=le.fit_transform(df.temp)
df.RH=le.fit_transform(df.RH)
```

```python
df.shape
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

# Spliting into independent & dependent variables

```python
#independent variable
x=df.iloc[:,:7]
x.head()
```

```python
#dependent variable
y=df.iloc[:,7:]
y.head()
```

```python
ss=StandardScaler()
x=ss.fit_transform(x)
y=ss.fit_transform(y)
```

```python
x.shape
```

```python
y.shape
```

```python
x
```

```python
x=pd.DataFrame(x)
x.head()
```

```python
y
```

```python
y=pd.DataFrame(y)
y.head()
```

# Train,Test&Split the Data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
x_train
```

# Creating the model MUltiLinearRegression

```python
from sklearn.linear_model import LinearRegression
```

```python
mlr=LinearRegression()
mlr.fit(x_train,y_train)
```

```python
predict=mlr.predict(x_test)
```

```python
predict
```

```python
y_test.head()
```

```python
from sklearn.metrics import r2_score
r2_score(mlr.predict(x_test),y_test)
```

```python
r2_score(mlr.predict(x_train),y_train)
```
